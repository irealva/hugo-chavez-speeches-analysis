## Widget to display progress

```python
def log_progress(sequence, every=None, size=None, name='Items'):
    from ipywidgets import IntProgress, HTML, VBox
    from IPython.display import display

    is_iterator = False
    if size is None:
        try:
            size = len(sequence)
        except TypeError:
            is_iterator = True
    if size is not None:
        if every is None:
            if size <= 200:
                every = 1
            else:
                every = int(size / 200)     # every 0.5%
    else:
        assert every is not None, 'sequence is iterator, set every'

    if is_iterator:
        progress = IntProgress(min=0, max=1, value=1)
        progress.bar_style = 'info'
    else:
        progress = IntProgress(min=0, max=size, value=0)
    label = HTML()
    box = VBox(children=[label, progress])
    display(box)

    index = 0
    try:
        for index, record in enumerate(sequence, 1):
            if index == 1 or index % every == 0:
                if is_iterator:
                    label.value = '{name}: {index} / ?'.format(
                        name=name,
                        index=index
                    )
                else:
                    progress.value = index
                    label.value = u'{name}: {index} / {size}'.format(
                        name=name,
                        index=index,
                        size=size
                    )
            yield record
    except:
        progress.bar_style = 'danger'
        raise
    else:
        progress.bar_style = 'success'
        progress.value = index
        label.value = "{name}: {index}".format(
            name=name,
            index=str(index or '?')
        )
```

```python
urls = [1,2,3,4,5,6,1000]

for url in log_progress(urls, every=1):
    print(url)
    
```

## Filtering Stopwords

```python
from nltk.corpus import stopwords
spanish_stops = stopwords.words('spanish')
```

```python
import unicodedata

#Note: this also removes the ñ, not just accents
def strip_accents(s):
   return ''.join(c for c in unicodedata.normalize('NFD', unicode(s, 'utf8'))
                  if unicodedata.category(c) != 'Mn')
```

```python
# Removing accents from our stopwords list
for i, word in enumerate(spanish_stops):
    stop_without_accent = strip_accents(spanish_stops[i].encode('utf-8'))
    spanish_stops[i] = stop_without_accent
    
# Printing stopwords list
spanish_stops
```

```python
# Next for loop removes stopwords
# We are using a python list comprehension to remove the tokens from Emma (after lowercasing them!) that are stopwords
for i, speech in enumerate(speeches):
    speeches[i] = [token.lower() for token in speech if token.lower() not in spanish_stops]
```

## TF-IDF (Term Frequency, Inverse Document Frequency)

```python
# code example from Building Machine Learning Systems with Python (Richert & Coelho) 
# - modified slightly by Lynn

import math

def tfidf(t, d, D):
    tf = float(d.count(t)) / sum(d.count(w) for w in set(d))  # normalized
    # Note his version doesn't use +1 in denominator.
    idf = math.log( float(len(D)) / (len([doc for doc in D if t in doc])))
    return tf * idf


a, abb, abc = ["a"], ["a", "b", "b"], ["a", "b", "c"]  # try adding another c to the last doc!
D = [a, abb, abc]

print(tfidf("a", a, D))   # a is in all of them
print(tfidf("a", abc, D)) # a is in all of them
print(tfidf("b", abc, D)) # b occurs only once here, but in 2 docs
print(tfidf("b", abb, D)) # b occurs more frequently in this doc
print(tfidf("c", abc, D)) # c is unique in the doc set
```

```python
from nltk.corpus import stopwords
stopword_list = stopwords.words('spanish')

import collections
def clean_tokens(tokens, stopwords):
    import string
    """ Lowercases, takes out punct and stopwords and short strings """
    return [token.lower() for token in tokens if (token not in string.punctuation)and (token.lower() not in stopwords) and len(token) > 2]

def makeText(filename, stopwords):
    from nltk import Text
    with open(filename) as handle:
        text = handle.read()
    return Text(clean_tokens(nltk.word_tokenize(text.decode('ascii', 'ignore')), stopwords))

def makeTextCollection(files, stopwords=stopword_list):
    from nltk import TextCollection
    texts= [makeText(filename, stopwords) for filename in files]
    collection = TextCollection(texts)
    return collection, texts

# use the data for the vocab in a single doc for a wordcloud, for instance
def compute_tfidf_by_doc(coll, texts, filenames):
    tfidf_by_doc = collections.defaultdict(list)
    i = 0
    for text in log_progress(texts, every=1):
    #for i, text in enumerate(texts):
        for word in set(text.tokens):   # just use the words in this text
            tfidfscore = coll.tf_idf(word, text)
            tf = coll.tf(word, text) # is actually count / len(text)
            count = text.count(word)
            if tfidfscore:
                tfidf_by_doc[filenames[i]].append({
                    "word": word,
                    "tfidf": tfidfscore,
                    "tf": tf,
                    "count": count
                })
        i = i + 1
    return tfidf_by_doc
```

```python
# We need to make the text collection, then use it to compute the tf-idf for the words in the docs.
res = makeTextCollection(filelist)
coll = res[0]
texts = res[1]
```

```python
tfidfs = compute_tfidf_by_doc(coll, texts, filelist)
```

```python
# Sorting tfidf from highest to lowest for each text in a collection
number_texts = len(tfidfs) # find number of texts

for i in range(0, number_texts):
    # Here we're just getting the tfidf terms for each text (first the file at 0, then 1, then 2... up to 12)
    one_text = tfidfs[tfidfs.keys()[i]]  # the first filename is the first key... it contains a list of words and scores    
    sorted_list = sorted(one_text, key=lambda k: k['tfidf'], reverse=True) # sort the frequency list based on the tfidf score
    text_key = tfidfs.keys()[i]
    tfidfs[text_key] = sorted_list
```

## Finding Most Common Pairs of Words ("Bigrams") (One speech)

```python
speech_to_process = speeches[7]

from nltk.collocations import *
bigram_measures = nltk.collocations.BigramAssocMeasures()

word_fd = nltk.FreqDist(speech_to_process) # all the words
bigram_fd = nltk.FreqDist(nltk.bigrams(speech_to_process))
finder = BigramCollocationFinder(word_fd, bigram_fd)
scored = finder.score_ngrams(bigram_measures.likelihood_ratio) # a good option here, there are others:
#scored[0:50]
```
