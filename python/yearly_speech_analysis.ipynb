## Tokenizing Words (and remove new lines, correct quotes, make lower case)

```python
for i, speech in enumerate(speeches):
    speeches[i] = speeches[i].replace('\n', ' ') # Remove new lines
    speeches[i] = speeches[i].replace('\r', ' ')
    speeches[i] = speeches[i].replace('“', '') # Correct quotes
    speeches[i] = speeches[i].replace('``', '')
    speeches[i] = speeches[i].replace('"', '')
    speeches[i] = speeches[i].replace('`', '')
    speeches[i] = speeches[i].replace('“', '') # Correct quotes 
    speeches[i] = speeches[i].decode("utf-8").replace(u"\xbf", "").encode("utf-8")  
    speeches[i] = speeches[i].decode("utf-8").replace(u"\xa1", "").encode("utf-8")
    speeches[i] = speeches[i].decode("utf-8").replace(u"(\u2018|\u2019)", "").encode("utf-8")
    speeches[i] = speeches[i].decode("utf-8").replace(u"\u0009", "").encode("utf-8")
    speeches[i] = speeches[i].decode("utf-8").replace(u'``', "").encode("utf-8")
    speeches[i] = speeches[i].decode("utf-8").replace(u'`', "").encode("utf-8")
    speeches[i] = speeches[i].decode("utf-8").replace(u'\u201d', "").encode("utf-8") ## Correcting 'right double quotation mark'
    
    speeches[i] = speeches[i].decode('utf8').encode('ascii', errors='ignore')
    
    #speeches[i] = str(speeches[i])
    
    speeches[i] = speeches[i].lower() # Make all the text lowercase
    #speeches[i] = nltk.wordpunct_tokenize(speeches[i])
    speeches[i] = nltk.word_tokenize(speeches[i].decode('utf-8')) # creates unicode list
```

## Filtering Stopwords

```python
from nltk.corpus import stopwords
spanish_stops = stopwords.words('spanish')
```

```python
import unicodedata

#Note: this also removes the ñ, not just accents
def strip_accents(s):
   return ''.join(c for c in unicodedata.normalize('NFD', unicode(s, 'utf8'))
                  if unicodedata.category(c) != 'Mn')
```

```python
# Removing accents from our stopwords list
for i, word in enumerate(spanish_stops):
    stop_without_accent = strip_accents(spanish_stops[i].encode('utf-8'))
    spanish_stops[i] = stop_without_accent
    
# Printing stopwords list
spanish_stops
```

```python
# Next for loop removes stopwords
# We are using a python list comprehension to remove the tokens from Emma (after lowercasing them!) that are stopwords
for i, speech in enumerate(speeches):
    speeches[i] = [token.lower() for token in speech if token.lower() not in spanish_stops]
```

# Removing punctuation

```python
import string
string.punctuation
```

```python
# Next for loop removes punctuation
# We are using a python list comprehension to remove the tokens from Emma (after lowercasing them!) that are stopwords
for i, speech in enumerate(speeches):
    speeches[i] = [token.lower() for token in speech if token.lower() not in string.punctuation]
```

## Counting Word Frequencies

```python
from nltk import Text

# Storring each speech frequency in this new array
speeches_frequency = []

for i, speech in enumerate(speeches):
    cleantext = Text(speeches[i])
    speeches_frequency.append(cleantext.vocab())

# Print top 20 most frequent words
speeches_frequency[0].most_common()[0:5]
```

```python
# Another way to do this is with nltk.FreqDist, which creates an object with keys that are 
# the vocabulary, and values for the counts:

# nltk.FreqDist(clean)['sashed']
```

```python
# Store our frequency data in a text file
newpath = '../data/organized_by_year/'

for i, speech in enumerate(speeches_frequency):
    wordpairs = speech.most_common()
    filename = newpath + speechnames[i] + '.frequency.txt'#creating a new path for the file
    with open(filename, "w") as handle:
        for pair in wordpairs:
            handle.write(pair[0].encode("utf-8") + "," + str(pair[1]) + "\n")
```
